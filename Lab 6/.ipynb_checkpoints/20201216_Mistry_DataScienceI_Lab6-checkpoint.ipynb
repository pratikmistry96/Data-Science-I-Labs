```python
# Load appropriate libraries
import numpy as np # Functions for vector calculations
import matplotlib.pyplot as plt # Plotting functions
import seaborn as sns # Data frame plotting
import pandas as pd # Manipulation of data frames
import scipy.cluster.hierarchy as shc # Heirarchical clustering
from sklearn.preprocessing import StandardScaler # Scaling functions
from sklearn.decomposition import PCA # Dimension reduction
from sklearn.cluster import AgglomerativeClustering # Clustering function
```

## 1) Read in both the microarray data (NCI60_data.csv) and the labels of the cancer type (NCI60_labs.csv).

```python
nci60_data = pd.read_csv('data/NCI60_data.csv',index_col=0) # Load data
nci60_labs = pd.read_csv('data/NCI60_labs.csv',index_col=0) # Load labels
```

```python
colname = list() # Create a list for column names
for ii in np.arange(1,1+nci60_data.shape[1]): colname.append('X{0}'.format(ii)) # Label features X1-totalfeatures
nci60_data.columns = colname # Change column names 
```

```python
nci60_data # View data
```

```python
nci60_labs # View labels
```

## 2) We will first do a PCA analysis. Scale, perform PCA, and plot the variance explained and the cumulative variance explained by the PCs. How many PCs are produced in the analysis? Approximately how many PCs are required to explain 90% of the variance in the data?

```python
## Scale nci60 data
scaler = StandardScaler() # initiate scaling function
pca = PCA() # Initiate PCA

scaler.fit(nci60_data) # Fit the scaling 
nci60_data_scale = scaler.fit_transform(nci60_data) # Transform the data

pca.fit(nci60_data_scale) # Fit pca to transformed data
pca_exp_var = pca.explained_variance_ratio_ # Extract explained variance ratio
pca_cum_var = np.cumsum(pca_exp_var) # Take the cumulative sum

plt.figure(figsize=(8,6)) # Set figure size
plt.plot(np.arange(1,1+len(pca_exp_var)), pca_cum_var) # Plot pc number and cumulative explained variance
plt.xlabel('PC Number') # Label x
plt.ylabel('Cumulative Explained Variance') # label y
plt.title('Cumulative Explained Variance by PC Number') # Add title
plt.show() # Show plot
```

### Answer:
From the plot of cumulative explained variance by PC number, it looks like approximately 40-45 PCs are required to explain 90% of the variance in the data

## 3) Plot the scores on the first versus second PC and the scores on the first versus third PC. Color the scores by cancer type. Interpret your results.

```python
nci60_data_pc = pca.fit_transform(nci60_data_scale) # Fit and transform the pca data
nci60_pc_df = pd.DataFrame(nci60_data_pc) # Create a data frame

# Create a list to name columns for PC and add to PC data frame
pcname = list()
for ii in np.arange(1,1+nci60_pc_df.shape[1]): pcname.append('PC{0}'.format(ii))
nci60_pc_df.columns = pcname
```

```python
plt.figure(figsize=(8,6)) # Set figure size
fig_1 = sns.scatterplot(x='PC1',y='PC2',data=nci60_pc_df,hue=nci60_labs.loc[:,'x'],legend='full') # Plot pc1 and pc2 and color by label
fig_1.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1) # Move legend to left
plt.title('PC2 vs PC1 Scores') # Add title
plt.show()
```

### Answer:

From the scatterplot of PC2 vs PC2, there doesn't seem to be obvious clusters between cancer types. Potentially, the bottom right of the plot groups colon cancers together.

```python
# Perform same operation as above but use PC1 and PC3
plt.figure(figsize=(8,6))
fig_2 = sns.scatterplot(x='PC1',y='PC3',data=nci60_pc_df,hue=nci60_labs.loc[:,'x'],legend='full')
fig_2.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)
plt.title('PC3 vs PC1 Scores')
plt.show()
```

### Answer:

From this plot comparing PC3 vs PC1, it seems a little bit more clear than the previous plot. The middle of the plot groups the ovarian cancer groups together, with a little cluster for colon next to it.

This emphasizes how difficult it is to visualize certain principal components with just two dimensions when the data contains many dimensions.

## 4) Next we will perform clustering on the data. Visualize the dendogram for hierarchical clustering with complete, average, and single linkage. Label the dendogram with the cancer types. Does the choice of linkage impact the results?

```python
plt.figure(figsize=(14,12)) # Set figure size
# Create dendrogram with euclidean distance and complete linkage
shc.dendrogram(shc.linkage(nci60_data_scale, 
                           metric = 'euclidean', 
                           method = 'complete'),
               orientation='right',
               leaf_font_size=8,
               labels = np.array(nci60_labs.loc[:,'x']))
plt.show()
```

Dendrogram using euclidean metric and complete linkage, and it appears to work well. With a few crossovers in certain groups.

```python
plt.figure(figsize=(14,12))
# Create dendrogram with euclidean distance and single linkage
shc.dendrogram(shc.linkage(nci60_data_scale, 
                           metric = 'euclidean', 
                           method = 'single'),
               orientation='right',
               leaf_font_size=8,
               labels = np.array(nci60_labs.loc[:,'x']))
plt.show()
```

Dendrogram with euclidean metric and single linakge. This method doesn't work as well as the complete linkage.

```python
plt.figure(figsize=(14,12))
# Create dendrogram with euclidean distance and average linkage
shc.dendrogram(shc.linkage(nci60_data_scale, 
                           metric = 'euclidean', 
                           method = 'average'),
               orientation='right',
               leaf_font_size=8,
               labels = np.array(nci60_labs.loc[:,'x']))
plt.show()
```

Finally, a dendrogram using euclidean distance with average linkage. It performs better than single linkage, but not as well as the complete linkage.

### Answer:

The linkage does affect the efficacy of the heirarchical clustering method and the dendrogram visualization. This emphasizes the importance of choosing the appropriate method when doing an analysis to avoid errors of clustering to ground truth data. From this example, it looks like the method using euclidean distance and complete linkage works the best when comparing the results to ground truth.

## 5) Use complete linkage for the remainder of the analysis. Cluster the observations into 4 groups. Look at the labels for each of the groups and comment on the success of the clustering.

```python
# Initiate agglomerative clustering
agglo_cluster = AgglomerativeClustering(n_clusters=4,
                                        affinity='euclidean',
                                        linkage='complete')
agglo_cluster.fit(nci60_data_scale) # Fit the cluster algorithm to data
agglo_cluster_pred = agglo_cluster.fit_predict(nci60_data_scale) # Fit and predict clusters

cluster_comp = pd.DataFrame() # Create a data frame
cluster_comp['Cancer Type'] = nci60_labs.loc[:,'x'] # Add ground truth labels
cluster_comp['Prediction'] = agglo_cluster_pred # Add predicted clusters
pd.crosstab(cluster_comp['Cancer Type'],
            cluster_comp['Prediction']) # Create a cross tab
```

### Answer:

The clustering works relatively well with a little "cross-talk" between the 4 clusters. Certain cancers were perfectly grouped into a group, such as NSCLC, Leukemia, Ovarian, and Prostate. However, there are a few cancers than have instances within other clusters, such as breast cancer, cns, and colon. Overall, it worked pretty well.

## 6) Rather than performing hierarchical clustering on the entire data matrix, we can simply perform hierarchical clustering on the first few principal component score vectors. Sometimes performing clustering on the first few principal component score vectors can give better results than performing clustering on the full data. Perform hierarchical clustering on the first 5 principal component scores and comment on the results.

```python
plt.figure(figsize=(14,12))
# Create a dendrogram using the top 5 PCs with euclidean distance and complete linkage
shc.dendrogram(shc.linkage(nci60_pc_df.iloc[:,0:5], 
                           metric = 'euclidean', 
                           method = 'complete'),
               orientation='right',
               leaf_font_size=8,
               labels = np.array(nci60_labs.loc[:,'x']))
plt.show()
```

```python
# Intiate a clustering instance using 4 groups
agglo_cluster = AgglomerativeClustering(n_clusters=4,
                                        affinity='euclidean',
                                        linkage='complete')

agglo_cluster.fit(nci60_pc_df.iloc[:,0:5]) # Fit algorithm to data
agglo_cluster_pred = agglo_cluster.fit_predict(nci60_pc_df.iloc[:,0:5]) # Fit and predict data

cluster_comp = pd.DataFrame() # Create a data frame
cluster_comp['Cancer Type'] = nci60_labs.loc[:,'x'] # Add ground truth label
cluster_comp['Prediction PCA'] = agglo_cluster_pred # Add predicted clusters
pd.crosstab(cluster_comp['Cancer Type'],
            cluster_comp['Prediction PCA'])
```

### Answer: 

Lastly, when only using the top 5 principal components after reducing the data and looking at the dendogram and the clustering method usign 4 groups. The principal components work well, but not perfect. It's important to remember that 5 principal components only capture 20% of the variance, so it's not ideal. We may be losing some information using that as the cutoff.
